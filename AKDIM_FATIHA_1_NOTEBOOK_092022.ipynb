##### 1.2.1.6. Clés primaires

Elle permet d'identifier de façon unique chaque enregistrement d'une table, mettre en relations avec d'autres tables et éviter les doublons.

METHODOLOGIE: On definit une fonction qui nous permettra de verifier si nous avons une clé primaire (en mentionnant la ou les colonnes utilisées). Le role de la fonction est de comparer la taille du df avec la taille du df sans les duplicata (qui ont été drop). Si les tailles correspondent alors nous avons une clé primaire.

```python
#Compter le nombre de 'client_id'
len(df_client['client_id'].unique())
```

```python
# on definit la fonction 
len(df_client['client_id'].unique())

def clesprimaire(df, colonnes):
    if df.size == df.drop_duplicates(colonnes).size :
        print("La clé n'est pas présente plusieurs fois dans le dataframe.")
        print("Elle peut donc être utilisée comme clé primaire.")
    else :
        print("La clé est présente plusieurs fois dans le dataframe.")
        print("Elle ne peut donc pas être utilisée comme clé primaire.")
    
    # On renvoie également le comptage
    print("Le dataframe est de la forme : " + str(df.shape) + " (lignes, colonnes)")
```

```python
#on applique cette fonction à notre
#df_client sur la colonne client_id
clesprimaire(df_client,'client_id')
```

**= > la donnée 'client_id' est utilisable en tant que clés primaire**

```python
#on crée une fonction qui va rechercher les outliers
def find_all_outliers(v):
    Q1 = np.quantile(v, 0.25)
    Q3 = np.quantile(v, 0.75)
    EIQ = Q3-Q1
    LI = Q1 - (EIQ*1.5)
    LS = Q3 + (EIQ*1.5)
    i = list(v.index[(v < LI)] | ( v >LS))
    val = list(v[i])
    return i, val
```

```python
#on applique cette fonction à notre df au niveau de la colonne 'price'
OUTLIER= find_all_outliers(df_produit['price'])
OUTLIER[1]
```

```python
#on tri ensuite par ordre décroissant l'extraction des données
OUTLIER[1].sort(reverse = True)
print(OUTLIER[1])
```

```python
#boxplot peu lisible on complète avec un zscore
outliers=[]
def detect_outliers(data):
    threshold=1
    mean=np.mean(data)
    std=np.std(data)
        
    for i in data:
        z_score=(i-mean) /std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

### 4.5. Inégalité de la distribution du chiffre d'affaires: Courbe de Lorenz

#### 4.5.1. Définition

```python
#c'est une représentation graphique de la distribution d'une variable 
#(Ici la variable est le Chiffre d'affaire)
#au sein d'un groupe de personne
#elle permet de calculer le coefficient de GINI qui mesure l'inégalité de la distribution
#de cette variable
#la droite d'equidistribution représente une situation dans laquelle la distribution de la variable
#se ferait de manière parfaitement égalitaire
```

#### 4.5.2. Courbe de lorenz

```python
variable = dfall2[dfall2['CA'] > 0]
var = variable['CA'].values
n = len(var)
lorenz = np.cumsum(np.sort(var)) / var.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0

plt.axes().axis('equal')
plt.tick_params(axis='both',which='major', labelsize=18)
#Il y a un segment de taille n pour chaque individu,
#plus 1 segment supplémentaire d'ordonnée 0.
#Le premier segment commence à 0-1/n, et le dernier termine à 1+1/n.
xaxis = np.linspace(0-1/n,1+1/n,n+1)
plt.plot(xaxis,lorenz,drawstyle='steps-post')
ident = [0.0,1.0]
plt.plot(ident,ident)
plt.fill_between(np.linspace(0,1, len(lorenz)), lorenz)
plt.show()
```

#### 4.5.3. Coef de Gini

```python
#calcul de l'indice de GINI
AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz.
#Le premier segment (lorenz[0]) est à moitié en dessous de 0,
#on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1]
#qui est à moitié au dessus de 1.
S = 0.5 - AUC # surface entre la première bissectrice et le courbe de Lorenz
gini = 2*S
gini
```

#### 4.5.4. Représentation

```python
variable = dfall2[dfall2['CA'] > 0]
var = variable['CA'].values
n = len(var)
lorenz = np.cumsum(np.sort(var)) / var.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0

plt.axes().axis('equal')
plt.text(0.2,0.08,"Gini="+str(gini))
plt.tick_params(axis='both',which='major', labelsize=18)
#Il y a un segment de taille n pour chaque individu,
#plus 1 segment supplémentaire d'ordonnée 0.
#Le premier segment commence à 0-1/n, et le dernier termine à 1+1/n.
xaxis = np.linspace(0-1/n,1+1/n,n+1)
plt.plot(xaxis,lorenz,drawstyle='steps-post')
ident = [0.0,1.0]
plt.plot(ident,ident)
plt.fill_between(np.linspace(0,1, len(lorenz)), lorenz)
plt.show()
```

```python
# Table de contingence
#la table de contingence résume sous forme de tableau les données
X = "categ"
Y = "sex"
cont = dfstat[[X, Y]].pivot_table(index=X, columns=Y, aggfunc=len).fillna(0).copy()
tx = dfstat[X].value_counts()
ty = dfstat[Y].value_counts()
cont = cont.astype(int)
cont
```

* En cas de relation, les dénombrements observés et attendus seront différents.

##### 5.2.2.3.Test non paramétrique de Spearman

##### 1- Explications

La corrélation de Spearman est l'équivalent non-paramétrique de la corrélation de Pearson. Elle mesure le lien entre deux variables. Si les variables sont ordinales, discrètes ou qu'elles ne suivent pas une loi normale, on utilise la corrélation de Spearman. Cette corrélation n'utilise pas les valeurs des données mais leur RANG. L'interprétation du coefficient de corrélation obtenu reste la même que lorsqu'on utilise une corrélation de Pearson. Le coefficient de corrélation varie entre -1 et +1, 0 reflétant une relation nulle entre les deux variables, une valeur négative (corrélation négative) signifiant que lorsqu'une des variable augmente, l'autre diminue ; tandis qu'une valeur positive (corrélation positive) indique que les deux variables varient ensemble dans le même sens

 Alors que le coefficient de corrélation de Pearson est une mesure de la relation linéaire entre deux variables, le coefficient de corrélation de Spearman mesure la relation monotone entre une paire de variables.

IL EXISTE DES FONCTIONS MONOTONES CROISSANTES, MONOTONES DÉCROISSANTES ET NON MONTONIQUES :

Pour une fonction monotone croissante, à mesure que X augmente, Y augmente également (et il n’est pas nécessaire qu’il soit linéaire). Pour une fonction monotone décroissante, à mesure qu’une variable augmente, l’autre diminue (elle ne doit pas non plus être linéaire). Une fonction non monotone est une fonction dans laquelle l’augmentation de la valeur d’une variable peut parfois entraîner une augmentation et parfois une diminution de la valeur de l’autre variable

Mesures du coefficient de corrélation de rang de Spearman the monotonic relation between two variables. Ses valeurs vont de -1 à +1 et peuvent être interprétées comme :

+1: Relation parfaitement monotone croissante
+0.8: Forte relation monotone croissante
+0.2: Relation croissante monotone faible
0: Relation non monotone
-0.2: Relation décroissante monotone faible
-0.8: Forte relation monotone décroissante
-1: Relation décroissante parfaitement monotone

##### 3 - Test de Kolmogorov-Smirnov

Le test mesure la distance entre la fonction de distribution empirique de l’échantillon et la fonction de distribution cumulative de la distribution de référence

**LES ETAPES:**
* 1 - Créér un tableau de valeurs à partir d’une distribution normale avec la moyenne et l’écart-type de nos données 
* 2 - Trier les données
* 3 - Calculer les sommes cumulées
* 4 - Tracez les fonctions de distribution cumulatives
  

```python
#Etape 1 : 
data_norm = np.random.normal(np.mean(df2['age']), np.std(df2['age']), len(df2))
#Etape 2 : 
values, base = np.histogram(df2['age'])
values_norm, base_norm = np.histogram(data_norm)
#Etape 3 : 
cumulative = np.cumsum(values)
cumulative_norm = np.cumsum(values_norm)
#Etape 4 : 
plt.plot(base[:-1], cumulative, c='blue')
plt.plot(base_norm[:-1], cumulative_norm, c='green')
plt.show()
```

* **ligne bleue**= fonction de distribution cumulative empirique de notre df
* **ligne verte**= distribution normale

```python
#methode seaborn
import seaborn as sns

sns.ecdfplot(df2['age'], c='blue')
sns.ecdfplot(data_norm, c='green')
```

*Le test est basé sur la distance maximale entre ces deux courbes (bleu – vert) avec les hypothèses suivantes:*

* *H0: deux échantillons proviennent de la même distribution*
* *H1: deux échantillons proviennent de distributions différentes*

```python
from scipy.stats import kstest

result = (kstest(df2['age'], cdf='norm'))

print(f"K-S statistic: {result[0]}")
print(f"p-value: {result[1]}")
```

**INTERPRETATION:**
La p value est égale à 0 et conforte le résultat rejet de l'hypothèse nulle les données ne sont pas distribuées normalement
